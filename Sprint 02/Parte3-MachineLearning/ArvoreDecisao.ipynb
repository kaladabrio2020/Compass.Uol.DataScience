```python
import matplotlib.pyplot as plt
from sklearn.tree     import DecisionTreeClassifier, export_graphviz, plot_tree
from sklearn.datasets import load_iris, make_moons
```

```python
iris = load_iris()
X, y = iris.data[:,2:], iris.target
```

```python
tree = DecisionTreeClassifier(max_depth=2)
tree.fit(X, y)
```

```python
export_graphviz(
    tree,
    out_file     ='iris.dot',
    feature_names=iris.feature_names[2:],
    class_names  =iris.target_names,
    rounded=True,
    filled =True
)
```

```python
plot_tree(tree, feature_names=iris.feature_names[2:], class_names=iris.target_names, rounded=True, filled=True)
plt.show()
```

```python
from plot.PlotRegiaDeDecisao import plotDecisionRegion
plotDecisionRegion(X, y,tree)
```

```python
X, y = make_moons(n_samples=100, noise=0.1)
tree2 = DecisionTreeClassifier()
tree2.fit(X,y)
plotDecisionRegion(X,y,tree2)
```

Restringir a arvore de decisão reduz o overfiting

```python
X, y = make_moons(n_samples=100, noise=0.2)
tree2 = DecisionTreeClassifier(min_samples_leaf=3)
tree2.fit(X,y)
plotDecisionRegion(X,y,tree2)
```

### Regressão

```python
from sklearn.datasets import make_regression
from sklearn.tree     import DecisionTreeRegressor
```
