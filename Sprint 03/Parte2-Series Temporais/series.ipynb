```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv(r'dataset/AirQualityUCI.csv',sep=';')[['Date','Time','CO(GT)','RH']]
```

```python
data.head(3)
```

```python
data.dropna(inplace=True)
```

tratamento dos dados

```python
data['Time'] = data['Time'].astype(str).str.replace('.',':')
```

```python
data['date_time'] = data['Date'] +' '+data['Time']
```

```python
data['date_time'] = pd.to_datetime(data['date_time'], format='mixed')
```

para CO(GT)

```python
data['CO(GT)'] = data['CO(GT)'].str.replace(',','.')
data['CO(GT)'] = data['CO(GT)'].astype(float)
```

para RH

```python
data['RH'] = data['RH'].str.replace(',','.')
data['RH'] = data['RH'].astype(float)
```

```python
data.groupby(by=['Date'])['RH'].mean().plot()
```
