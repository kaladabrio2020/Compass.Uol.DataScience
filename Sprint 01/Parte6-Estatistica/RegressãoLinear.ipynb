## Regressão Linear Simples


$$Ŷ = \beta_0 + \beta_1*X_i$$
* $Ŷ$ : valor resposta (target)
* $X_i$ : valor explicativa (feature)
* $\beta_0$: coeficiente linear (interceptor)
* $\beta_1$: coeficiente angular (peso, coef_, )



##### Coef linear
$\bar{\beta_0} = \bar{Y} - \beta_0 * \bar{X}$


##### Coef Angular
$\bar{\beta_1} = \bar{Y} - \beta_0 * \bar{X}$

```python
data = pd.read_csv('dataset/idadePesos.csv',sep=',')
data.head(3)
```

```python
plt.scatter(x=data['peso'],y=data['idade'])
plt.show()
```

```python
data.corr(method='pearson')
```

```python
from scipy.stats import linregress

coef_ , intercept = linregress(x=data['peso'],y=data['idade'])[0:2]
```

```python
f = lambda x: intercept + coef_ * x
plt.scatter(x=data['peso'],y=data['idade'])
plt.plot(data['peso'],f(data['peso'].values),c='r')
plt.show()
```

```python
from sklearn.metrics import r2_score, mean_absolute_error
r2_score(data['idade'],f(data['peso'].values)),mean_absolute_error(data['idade'],f(data['peso'].values))
```
