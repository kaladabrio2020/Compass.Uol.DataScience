```python
import pandas as pd
import numpy  as np
from scipy.stats import binom
```

```python
data = pd.read_csv(r'dataset/example.csv',sep=',')
df1 = data['Numero'].value_counts().reset_index()
```

# Distribuiçao de Probabilidades

```python
# Probabilidade de um evento ocorrer
df1['P(Xi)'] = df1['count']/df1['count'].sum()
df1
```

```python
# Esperança de um valor
df1['(Xi*P(Xi))'] = df1['Numero']*df1['P(Xi)']
display(df1)
display(df1['(Xi*P(Xi))'].mean())
```

```python
df1['std'] = np.power(( df1['Numero'] * df1['(Xi*P(Xi))'] ),2) * df1['P(Xi)']
display(df1)
display(df1['std'].mean())
```
