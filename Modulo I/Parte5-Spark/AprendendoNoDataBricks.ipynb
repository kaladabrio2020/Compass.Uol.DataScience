## Agrupamento

```
data = [("Anderson","Sales","NY",90000,34,10000),
    ("Kenedy","Sales","CA",86000,56,20000),
    ("Billy","Sales","NY",81000,30,23000),
    ("Andy","Finance","CA",90000,24,23000),
    ("Mary","Finance","NY",99000,40,24000),
    ("Eduardo","Finance","NY",83000,36,19000),
    ("Mendes","Finance","CA",79000,53,15000),
    ("Keyth","Marketing","CA",80000,25,18000),
    ("Truman","Marketing","NY",91000,50,21000)
  ]
schema = ["emp_name","dep_name","state","salary","age","bonus"]
data = spark.createDataFrame(data=data, schema=schema)
data.show(2)
```

```
data.printSchema()
data.show(truncate=False)
```

```
data.groupBy(F.col('state')).avg('bonus').show()
```

```
data.groupBy(F.col('state')).count().show()
```

```
data.groupBy(F.col('state'),F.col('dep_name')).mean().show(truncate=False)
```

```
data.groupBy(F.col('state'))\
    .agg(  
        F.mean('age').alias('mediaIdade')
        ).show(3)
```

```
data.groupBy(F.col('dep_name'))\
    .agg(
        F.mean('salary').alias('salario'),
        F.stddev('age').alias('dsvioPadra')
    ).filter(
        F.col('salario')>87000
    ).show()
```

```
data.select('*')\
    .filter(
        F.col('state')=='NY'
        ).show()
    
data.select('*')\
    .where(
        F.col('state')=='NY'
    ).show()
```

```
data.select('salary').distinct().count()
```

```
data:dataframe = (
    spark.read
    .format("csv")
    .option('header',"True")
    .option('inferschema','True')
    .option("delimiter",';')
    .load('/FileStore/tables/arquivo_geral-2.csv')
)
display(data)
```

```
s = data.select(F.col('regiao'),F.col('data'),F.col('casosNovos'))\
    .where(
        (F.col('data') > F.lit('2020-02-01')) &
        (F.col('data') < F.lit('2020-02-27'))
    )\
    .groupBy('regiao')\
        .count() 
display(s)
```

## Join

```
data = [
    (1, "Anderson", 10000),
    (2, "Kenedy", 20000),
    (3, "Billy", 23000),
    (4, "Andy", 23000),
    (5, "Mary", 24000),
    (6, "Eduardo", 19000),
    (7, "Mendes", 15000),
    (8, "Keyth", 18000),
    (9, "Truman", 21000),
]

schema = ["id", "name", "salary"]
df1 = spark.createDataFrame(data=data, schema=schema)

data = [
    (1, "Delhi", "India"),
    (2, "Tamil nadu", "India"),
    (3, "London", "UK"),
    (4, "Sydney", "Australia"),
    (8, "New York", "USA"),
    (9, "California", "USA"),
    (10, "New Jersey", "USA"),
    (11, "Texas", "USA"),
    (12, "Chicago", "USA"),
]

schema = ["id", "place", "country"]
df2 = spark.createDataFrame(data=data, schema=schema)
```

```
# Inner Join

df1.join(other=df2, on=['id'], how='inner').show()
```

```
df1.join(other=df2, on=['id'], how='left').show()
```
